```python
import pandas as pd
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as o
from sklearn import preprocessing
```

```python
df_studio = pd.read_excel(r'tabela_aluguel_studios.xlsx')
df_studio.head()
```

```python
df_studio['Zona'].unique()
```

```python
df_studio.shape
```

```python
# Removendo coluna inútil

df_studio = df_studio.drop(columns = ['Unnamed: 0'])
display(df_studio)
```

```python
# Removendo valores com 0, caso haja
df_studio = df_studio[(df_studio.Zona != 0) & (df_studio.Bairro != 0) & (df_studio.Dormitorios != 0)]
display(df_studio)
```

```python
df_studio.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
df_studio.describe()
```

```python
df_studio['Banheiros'].describe()
```

```python
df_studio['Banheiros'].value_counts()
```

```python
ban_ = df_studio[df_studio['Banheiros'] != 1]
ban_
```

```python
df_studio = df_studio.drop(ban_.index, axis=0)
```

```python
df_studio.describe()
```

```python
suitex = df_studio[(df_studio['Suites'] != 1) & (df_studio['Suites'] != 0)]

suitex
```

```python
df_studio = df_studio.drop(suitex.index, axis=0)
df_studio.describe()
```
