```python
cols = df_studio.columns 
colours = ['#000099', '#ffff00'] # amarelo são missing values e azul são valores presentes
plt.figure(figsize=(10,8))
sns.heatmap(df_studio[cols].isnull(), cmap=sns.color_palette(colours))
```

```python
# Varificando porcentagem de missing values 
for col in df_studio.columns:
    pct_missing = np.mean(df_studio[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Removendo os valores 0

df_studio = df_studio[(df_studio.Zona != 0) & (df_studio.Bairro != 0) & (df_studio.Dormitorios != 0)]
display(df_studio)
```

```python
df_studio['Zona'].value_counts()
```

```python
# Tentando definir livremente  uma quantidade razoável para as colunas Suites, Vagas e Banheiros
# Pois apresentam muitos valores absurdos

trial = df_studio[(df_studio.Banheiros <= 3) & (df_studio.Vagas <= 2) & (df_studio.Suites <= 1)]
display(trial)
```
