```python
import pandas as pd
import numpy as np
```

```python
dff = pd.read_csv(r'table_housi_ofertas.csv', low_memory=False)
display(dff)
```

```python
dff.columns
```

```python
df_rent = dff[['Id_Oferta', 'Portal', 'Bairro', 'Zona', 'Cidade', 'Status', 'Valor_Iptu', 'Valor_Condominio', 'Valor_Aluguel', 'Area_Privativa', 'Dormitorios', 'Suites', 'Vagas', 'Banheiros']]
display(df_rent)
```

```python
df_rent['Aluguel_m2'] = df_rent['Valor_Aluguel'] / df_rent['Area_Privativa']
df_rent['Aluguel_m2'] = df_rent['Aluguel_m2'].apply(lambda x: round(x,2))

df_rent
```

```python
n_imov = list(df_rent['Bairro'].value_counts())

print(n_imov)
```

```python
bairros = list(df_rent['Bairro'].unique())
print(bairros)
```

```python
print(len(n_imov))
print(len(bairros))
```

```python
# Análise estatística dos bairros e total de imóveis

df_bairros = pd.DataFrame(data=bairros, columns=['Bairros'])
display(df_bairros)
```

```python
df_imov = pd.DataFrame(data=n_imov, columns=['Imoveis'])
display(df_imov)
```

```python
df_anal = pd.concat([df_bairros, df_imov], axis=1)
display(df_anal)
```

```python
main = []
for bairro in list(df_rent['Bairro'].value_counts()):
    if bairro > 1000:
        main.append(bairro)
print(main)
```

```python
df_rent['Bairro'].count()
```

```python
borough = []
for bairro, qtd in enumerate(n_imov):
    if qtd > 1000:
        if bairros[bairro] in borough:
            pass
        else:
            borough.append(bairros[bairro])
print(borough)
```

```python
dados = pd.DataFrame({'Bairros': borough, 'Imoveis': main})
df_principais = pd.DataFrame(data=dados)

display(df_principais)
```

```python
# Selecionar apenas os bairros acima de 1000 imóveis no DF original.

df_fltd = df_rent.loc[(df_rent['Bairro']).isin(df_principais['Bairros'])]
df_fltd
```
